```python


def calculate_distance(zip1, zip2):
    """Calculate distance between two zip codes using pgeocode."""
    try:
        # Convert zip codes to strings and remove any decimal points
        zip1 = str(int(float(zip1))) if pd.notna(zip1) else None
        zip2 = str(int(float(zip2))) if pd.notna(zip2) else None

        if zip1 is None or zip2 is None:
            return np.nan

        # Initialize pgeocode for US
        nomi = pgeocode.Nominatim('us')

        # Get coordinates
        data1 = nomi.query_postal_code(zip1)
        data2 = nomi.query_postal_code(zip2)

        # Calculate distance if coordinates are found
        if data1.latitude is not np.nan and data1.longitude is not np.nan and data2.latitude is not np.nan and data2.longitude is not np.nan:
            distance_km = nomi.distance_between_postal_codes(zip1, zip2)
            distance_miles = distance_km * 0.621371  # Convert to miles
            return distance_miles
        else:
            return np.nan
    except:
        return np.nan

```

```python
from geopy.distance import geodesic

def get_lat_long(zip_code):
    location = nomi.query_postal_code(zip_code)
    return location.latitude, location.longitude


def get_distance(zip1, zip2):
    lat1, long1 = get_lat_long(zip1)
    lat2, long2 = get_lat_long(zip2)
    return geodesic((lat1, long1), (lat2, long2)).miles

```
